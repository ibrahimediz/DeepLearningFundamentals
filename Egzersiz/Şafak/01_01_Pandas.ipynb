```python
def pct10(column):
    if str(column.dtype) == "object":
        return 0
    else:
        return column.quantile(0.10)
def pct90(column):
    if str(column.dtype) == "object":
        return 0
    else:
        return column.quantile(0.90)
resault.agg([pct10,pct90])
```
