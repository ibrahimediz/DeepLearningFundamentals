```python
import pandas as pd
df = pd.read_csv("/workspace/DeepLearningFundamentals/preprocessyeni.csv")
df.head()
```

```python
df["Sleep Disorder"].unique()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="constant",fill_value="Healthy")
imp = pd.DataFrame(data=imputer.fit_transform(df[["Sleep Disorder"]]))
df["Sleep Disorder"] = imp[0]
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[("encode",OneHotEncoder(),[0,2,7])],remainder="passthrough")
dfEncoded = pd.DataFrame(ct.fit_transform(df))
dfEncoded.head()
```
