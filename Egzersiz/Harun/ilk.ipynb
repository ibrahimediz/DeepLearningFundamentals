```python
import pandas as pd
df = pd.read_csv("/workspace/DeepLearningFundamentals/preprocess1.csv")
```

```python
df2 = df.copy()
df3 = df2.dropna(axis=1) # sütun silme
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
df3["status"]=pd.DataFrame(data=enc.fit_transform(df3[["status"]]).toarray())

df3.head()
```

```python
df4 = df3.drop("org_title", axis=1 )
#df5 = df4.drop("hours", axis=1)
df6 = df4.drop("display_url", axis=1)
df7 = df6.drop("region", axis=1)
df8 = df7.drop("last_modified_date", axis=1)
df9 = df8.drop("created_date", axis=1)
df9.head()
```

```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df9['recurrence_type']= label_encoder.fit_transform(df9['recurrence_type'])
df9.head()
```
