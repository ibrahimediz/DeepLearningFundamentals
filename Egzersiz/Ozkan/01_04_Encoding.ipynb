```python
adres7 = "/workspace/DeepLearningFundamentals/preprocess1.csv"
df7 = pd.read_csv(adres7,index_col=[0])
df7.head()
```

```python
df7.isna().sum()
```

```python
veri7 = df7.copy()
veri7=veri7.dropna(axis=1)
```

```python
veri7.isna().sum()
```

```python
veri7.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(veri7["org_title"])
veri7["EnOrg_title"] = le.transform(veri7["org_title"])
le.fit(veri7["recurrence_type"])
veri7["EnRecurrence_type"] = le.transform(veri7["recurrence_type"])

veri7.head()
```

```python
itr_Imp = IterativeImputer()
itr_Imp.fit_transform(veri7[["EnOrg_title","EnRecurrence_type"]])
sonuc = pd.DataFrame(data=itr_Imp.fit_transform(veri7[["EnOrg_title","EnRecurrence_type"]]))
sonuc
```
