```python
out=[]
def iqr_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    Lower_tail = q1 - 1.5 * iqr
    Upper_tail = q3 + 1.5 * iqr
    # for i in colunm:
    #     if i > Upper_tail or i < Lower_tail:
    #         out.append(i)
    return df[~((df<Lower_tail) | (df>Upper_tail)).any(axis=1)]
df_clean = iqr_outliers(df)
```
