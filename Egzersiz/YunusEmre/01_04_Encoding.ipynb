```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

adres = "/workspace/DeepLearningFundamentals/preprocess1.csv"
df = pd.read_csv(adres)
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.drop(['Latitude', 'Longitude', 'Census Tract', 'BIN', 'BBL', 
'NTA'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
imputer = SimpleImputer(strategy="most_frequent")
df["category_id"]=imputer.fit_transform(df[["category_id"]])
```

```python
df["postalcode"]=imputer.fit_transform(df[["postalcode"]])
```

```python
df.isna().sum()
```

```python
df.drop("org_title",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[0,1])],remainder="passthrough")
ct.fit_transform(df)
encoded=pd.DataFrame(data=ct.fit_transform(df))
encoded.drop([6],axis=1,inplace=True)
encoded.columns=list(range(10))
encoded.head()
```
