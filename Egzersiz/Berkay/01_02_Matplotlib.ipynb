```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
renkler = {"Yes":"b","No":"r"}
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="smoker",palette=renkler)
```

```python
renkler1 = {"Male":"b","Female":"r"}
sns.countplot(x="smoker",data=tips,hue="sex",palette=renkler1)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",row="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",row="smoker",col="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size="size")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size="size",alpha=0.4)
```
