```python

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
df_pre = df.copy()
le = LabelEncoder()
le.fit(df_pre["BMI Category"])
df_pre["BMI Category"] = le.transform(df_pre["BMI Category"])
le.fit(df_pre["Gender"])
df_pre["Gender"] = le.transform(df_pre["Gender"])
df_pre = df_pre.drop(["Person ID"],axis=1)
df_pre
```

```python
le.fit(df_pre["Occupation"])
df_pre["Occupation"] = le.transform(df_pre["Occupation"])
df_pre.head(5)
```

```python
df.isna().sum()
```

```python
df_pre['Blood Pressure Low'] = df_pre['Blood Pressure'].str.split('/').str[0]
df_pre['Blood Pressure High'] = df_pre['Blood Pressure'].str.split('/').str[1]
df_pre = df_pre.drop("Blood Pressure",axis=1)
df_pre.head(5)
```

```python
df["Occupation"].unique()
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values=np.NaN, strategy='constant', fill_value="Healthy")
imp = pd.DataFrame(imputer.fit_transform(df_pre[["Sleep Disorder"]]))
df_pre["Sleep Disorder"] = imp[0]
```

```python
le.fit(df_pre["Sleep Disorder"])
df_pre["Sleep Disorder"] = le.transform(df_pre["Sleep Disorder"])
df_pre.head(5)
```
