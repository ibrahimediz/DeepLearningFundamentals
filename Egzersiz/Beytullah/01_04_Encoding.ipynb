```python
#############################  3.
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(strategy="constant",fill_value="Healthy")
imp = pd.DataFrame(data=imputer.fit_transform(df[["Sleep Disorder"]]))
df["Sleep Disorder"] = imp[0]
df.head()
```

```python
df.isna().sum()
```

```python
#### 4
df.info()
```

```python
df["bldPrB"] = df["Blood Pressure"].apply(lambda x:x.split('/')[0])
df["bldPrS"] = df["Blood Pressure"].apply(lambda x:x.split('/')[1])
df.drop(["Person ID","Blood Pressure"],axis=1,inplace=True)
df.head()
```

```python
df.Occupation.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[("encode",OneHotEncoder(),[0,2,7])],remainder="passthrough")
dfEncoded = pd.DataFrame(ct.fit_transform(df))
dfEncoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[24] = le.fit_transform(dfEncoded[24])
```

```python
dfEncoded.to_csv("Cevap.csv",index=False)
```
